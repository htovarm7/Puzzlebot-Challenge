```python
model = YOLO("yolov8n.pt")  

results = model.train(
    data="Traffic-Signals-5/data.yaml",
    epochs=300,
    imgsz=256,            
    batch=16,
    patience=20,
    fliplr=0.0,
    flipud=0.0,
    degrees=0.0,
    shear=0.0,
    project="runs",
    name="train",
)
```

```python
metrics = model.val()
print(metrics.box.map)     # mAP50-95
print(metrics.box.map50)   # mAP50
print(metrics.box.maps)    # per-class mAP50-95
```
